# Data Cleaning & Preprocessing

## Handle Missing Value

```python
df.isna().sum()
```

```python
for col in ['daily_time_spent_on_site', 'area_income', 'daily_internet_usage']:
  df[col] = df[col].fillna(df[col].median())

df['male'] = df['male'].fillna(df['male'].mode()[0])
```

Feature yang memiliki missing values akan di-impute dengan nilai median nya kecuali feature male yang di-impute dengan mode nya.

## Handling Duplicate Data

```python
df.duplicated().sum()
```

## Extract Datetime

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['hour'] = df['timestamp'].dt.hour + (df['timestamp'].dt.minute/60).round(2)

df[['timestamp', 'year', 'month', 'day', 'hour']].sample(3)
```

## Feature Encoding

### Label Encoding

```python
df['male'] = np.where(df['male'] == 'Laki-Laki', 1, 0)
df['clicked_on_ad'] = np.where(df['clicked_on_ad'] == 'No', 0, 1)
```

### One Hot Encoder

```python
df['city_province'] = df['city'].astype(str)+ '-' +df['province'].astype(str)
ohe = pd.get_dummies(df[['city_province', 'category']])
df = df.join(ohe)
```

## Split Data

### Without Normalization

- Data Train 80
- Data Tes 20


```python
df_copy = df.copy()
X = df.drop(['timestamp', 'clicked_on_ad', 'city', 'province', 'category', 'city_province'], axis=1)
y = df['clicked_on_ad']

X_train_80, X_test_20, y_train_80, y_test_20 = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=1
)

print(X_train_80.shape, X_test_20.shape, y_train_80.shape, y_test_20.shape)
```

# Modeling

```python
model_list = {
    'Logistic Regression':LogisticRegression(max_iter=500, random_state=1),
    'KNN':KNeighborsClassifier(),
    'SVC':SVC(random_state=1),
    'Decision Tree':DecisionTreeClassifier(random_state=1),
    'Random Forest':RandomForestClassifier(random_state=1)
}

y_pred_list = dict()
```

```python
def get_score(y_pred_list, y_test):
  model_name = []
  accuracy = []
  precision = []
  recall = []
  f1 = []
  roc_auc = []

  for name, y_pred in y_pred_list.items():
    model_name.append(name)
    accuracy.append(accuracy_score(y_test, y_pred))
    precision.append(precision_score(y_test, y_pred))
    recall.append(recall_score(y_test, y_pred))
    f1.append(f1_score(y_test, y_pred))
    roc_auc.append(roc_auc_score(y_test, y_pred))

    score_list = {
        'model':model_name,
        'accuracy':accuracy,
        'precision':precision,
        'recall':recall,
        'f1_score':f1,
        'roc_auc':roc_auc
      }

  score_df = pd.DataFrame(score_list).set_index('model')

  return score_df
```

### Without Normalization

- Data Train 80
- Data Tes 20

```python
for name, model in model_list.items():
  model.fit(X_train_80, y_train_80)
  y_pred_list[name] = model.predict(X_test_20)

get_score(y_pred_list, y_test_20).round(2)
```

- Data Train 70
- Data Tes 30

```python
for name, model in model_list.items():
  model.fit(X_train_70, y_train_70)
  y_pred_list[name] = model.predict(X_test_30)

get_score(y_pred_list, y_test_30).round(2)
```

### WIth Normalization

- Data Train 80
- Data Tes 20

```python
for name, model in model_list.items():
  model.fit(X_train_scaled_80, y_train_scaled_80)
  y_pred_list[name] = model.predict(X_test_scaled_20)

get_score(y_pred_list, y_test=y_test_scaled_20).round(2)
```

- Data Train 70
- Data Test 30

```python
for name, model in model_list.items():
  model.fit(X_train_scaled_70, y_train_scaled_70)
  y_pred_list[name] = model.predict(X_test_scaled_30)

get_score(y_pred_list, y_test=y_test_scaled_30).round(2)
```

**Observation**
- Model terbaik dengan accuracy score 97% adalah Logistic Regression dengan komposisi data train 80% dan data test 20% setelah dilakukan normalisasi

```python
model = LogisticRegression(max_iter=500, random_state=1)
model.fit(X=X_train_scaled_80, y=y_train_scaled_80)
y_pred = model.predict(X_test_scaled_20)
print(classification_report(y_test_scaled_20, y_pred))
```
